## 4 Pré-processamento + Baseline com Vetorização (TF-IDF n-gram) e Modelo supervisionado

```python
# Padronização/limpeza de texto
def basic_clean(text):
    """
    Realiza o pré-processamento básico de texto para análise de NLP.
    
    Esta função implementa técnicas fundamentais de normalização textual que são
    essenciais para melhorar a qualidade da representação dos dados textuais.
    O objetivo é reduzir a variabilidade superficial do texto mantendo seu
    conteúdo semântico, permitindo que os algoritmos de machine learning
    identifiquem padrões mais consistentes.
    
    Parâmetros:
    ----------
    text : str ou qualquer tipo
        O texto a ser processado. Será convertido para string se necessário.
    
    Retorna:
    -------
    str
        Texto processado e normalizado.
    
    Técnicas aplicadas:
    ------------------
    - Conversão para minúsculas: reduz dimensionalidade do vocabulário
    - Remoção de acentos: normaliza caracteres especiais do português
    - Remoção de URLs: elimina ruído não semântico
    - Remoção de pontuação: foca no conteúdo lexical
    - Normalização de espaços: padroniza separação de tokens
    """
    if not isinstance(text, str):
        text = str(text)
    text = text.lower()
    text = unidecode(text)
    text = re.sub(r"http\S+|www\.\S+", " ", text)
    text = re.sub(r"[^a-z0-9\s]", " ", text)
    text = re.sub(r"\s+", " ", text).strip()
    return text
```

```python
X = df['descricao_reclamacao'].astype(str).apply(basic_clean)
y = df['categoria'].astype(str)
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.25, random_state=42, stratify=y
)
```

```python
baseline = Pipeline([
    ('tfidf', TfidfVectorizer(ngram_range=(1,2), min_df=2, max_df=0.9)),
    ('clf', LogisticRegression(max_iter=2000))
])
```

```python
# Treinando e avaliando o modelo baseline
baseline.fit(X_train, y_train)
pred = baseline.predict(X_test)
```

```python
# F1 Score
f1 = f1_score(y_test, pred, average='weighted')
print("F1 Score:", f1)

# Relatório de classificação
print("\nClassificação:")
print(classification_report(y_test, pred))
```

### Análise dos Resultados do Modelo Baseline

O modelo baseline utilizando TF-IDF com bigramas e Regressão Logística representa nossa primeira abordagem sistemática para o problema de classificação de chamados. Os resultados obtidos servirão como referência para avaliar o desempenho de modelos mais complexos desenvolvidos posteriormente.

**Justificativa da Abordagem:**
- **TF-IDF**: Escolhido por sua eficácia em representar a importância relativa dos termos nos documentos, penalizando palavras muito comuns e valorizando termos discriminativos.
- **N-gramas**: A inclusão de bigramas permite capturar algumas relações contextuais entre palavras adjacentes.
- **Regressão Logística**: Selecionada por sua simplicidade, interpretabilidade e eficiência computacional para problemas de classificação multiclasse.

## 5 Melhorando o modelo (TF + LinearSVC)"

```python
import nltk
nltk.download('stopwords')

# lista de stopwords do NLTK
stops = nltk.corpus.stopwords.words('portuguese')
```

```python
# Pipeline aprimorado: TF (sem IDF) + LinearSVC + stopwords
"""
Implementação de pipeline otimizado baseado nas limitações identificadas no modelo baseline.

Modificações implementadas:
1. TfidfVectorizer com use_idf=False: utiliza apenas Term Frequency, eliminando
   a ponderação por Inverse Document Frequency que pode não ser adequada para
   textos curtos como chamados de atendimento.

2. LinearSVC: Support Vector Machine linear, conhecido por sua eficácia em
   classificação de texto, especialmente quando há muitas features esparsas
   como no caso de representações TF-IDF.

3. stop_words: incorporação de stopwords do português para reduzir dimensionalidade
   e eliminar palavras funcionais sem valor semântico discriminativo.

Justificativa das escolhas:
- SVM é particularmente eficaz em espaços de alta dimensionalidade
- A remoção de stopwords reduz o ruído nos dados textuais
- TF sem IDF pode ser mais adequado para textos de domínio específico
"""
pipe_svc = Pipeline([
    ('tfidf', TfidfVectorizer(use_idf=False, stop_words=stops)),
    ('clf', LinearSVC())
])
```

```python
# Definição do grid de hiperparâmetros para otimização
"""
Grid Search para otimização sistemática de hiperparâmetros. A seleção dos
parâmetros foi baseada na literatura de NLP e nas características específicas
dos dados de chamados de atendimento.

Parâmetros explorados:
- ngram_range: comparação entre unigramas (1,1) e unigramas+bigramas (1,2)
  para avaliar o impacto do contexto local na classificação
- min_df: frequência mínima de documentos (1, 2, 3) para controle de vocabulário
- max_df: frequência máxima de documentos (0.9, 0.95) para filtragem automática
- C: parâmetro de regularização do SVM (0.5, 1.0, 2.0) controlando o trade-off
  entre margem de separação e erro de classificação

O Grid Search utilizará validação cruzada (cv=3) para avaliação robusta dos
hiperparâmetros, com F1-score ponderado como métrica de otimização.
"""
param_grid = {
    'tfidf__ngram_range': [(1,1), (1,2)],
    'tfidf__min_df': [1, 2, 3],
    'tfidf__max_df': [0.9, 0.95],
    'clf__C': [0.5, 1.0, 2.0],
}
```

```python
# Realizando o Grid Search com validação cruzada
grid = GridSearchCV(
    pipe_svc,
    param_grid=param_grid,
    scoring='f1_weighted',
    cv=3,
    n_jobs=-1,
    verbose=1
)
```

```python
# Treinando o modelo
grid.fit(X_train, y_train)

print("Melhor F1 (CV):", grid.best_score_)
print("Melhores parâmetros:", grid.best_params_)
```

```python
# Avaliando no conjunto de teste
best_model = grid.best_estimator_
pred_best = best_model.predict(X_test)

# Cálculo do F1 Score
f1_best = f1_score(y_test, pred_best, average='weighted')
print("\nF1 Score - TF-IDF + LinearSVC:", f1_best)
print("\nClassificação:")
print(classification_report(y_test, pred_best))
```

### Análise dos Resultados da Otimização com Grid Search

Os resultados obtidos através da otimização de hiperparâmetros demonstram a importância da sintonização sistemática de modelos de machine learning. O Grid Search permitiu identificar a combinação ótima de parâmetros através de validação cruzada, garantindo maior robustez na avaliação.

**Interpretação dos Resultados:**
- **Melhor F1 Score (CV)**: Representa a performance média obtida durante a validação cruzada no conjunto de treinamento
- **Melhores Parâmetros**: Configuração otimizada identificada pelo algoritmo de busca em grade
- **F1 Score no Conjunto de Teste**: Avaliação imparcial da capacidade de generalização do modelo

A diferença entre o F1-score da validação cruzada e o F1-score no conjunto de teste indica o nível de generalização do modelo e possível overfitting ou underfitting.

## 6 Modelo com Embeddings

```python
# Classificador baseado em embeddings densos
torch.set_num_threads(2)

# Funções auxiliares para Word2Vec Skip-gram + Negative Sampling
def build_vocab(texts, min_count=8, max_vocab=20000):
    """
    Constrói vocabulário a partir dos textos de treinamento.
    
    Parâmetros:
    ----------
    texts : list
        Lista de strings (textos limpos)
    min_count : int
        Frequência mínima para inclusão no vocabulário
    max_vocab : int
        Tamanho máximo do vocabulário
        
    Retorna:
    -------
    tuple
        (string_to_index, index_to_string, counter)
    """
    counter = Counter()
    for t in texts:
        counter.update(t.split())
    vocab = [w for w,c in counter.items() if c >= min_count]
    vocab = sorted(vocab, key=lambda w: counter[w], reverse=True)[:max_vocab-2]
    itos = ['<pad>', '<unk>'] + vocab
    stoi = {w:i for i,w in enumerate(itos)}
    return stoi, itos, counter

# Converte textos em índices
def texts_to_indices(texts, stoi):
    """Converte textos em sequências de índices."""
    unk = stoi['<unk>']
    return [[stoi.get(w, unk) for w in t.split()] for t in texts]

# Gera pares (center, context) para Skip-gram
def generate_skipgram_pairs(seqs, window=2, max_pairs=40000):
    """
    Gera pares skip-gram (palavra_central, palavra_contexto).
    
    Parâmetros:
    ----------
    seqs : list
        Lista de sequências de índices
    window : int
        Tamanho da janela de contexto
    max_pairs : int
        Número máximo de pares a gerar
        
    Retorna:
    -------
    np.ndarray
        Array de pares (center_word, context_word)
    """
    pairs = []
    for seq in seqs:
        n = len(seq)
        for i, center in enumerate(seq):
            start = max(0, i-window)
            end = min(n, i+window+1)
            for j in range(start, end):
                if j == i:
                    continue
                pairs.append((center, seq[j]))
                if len(pairs) >= max_pairs:
                    return np.array(pairs, dtype=np.int64)
    return np.array(pairs, dtype=np.int64)

# Modelo Skip-gram com Negative Sampling
class SkipGramNS(torch.nn.Module):
    """
    Modelo Skip-gram com Negative Sampling implementado em PyTorch.
    
    Arquitetura:
    - in_emb: embeddings para palavras centrais
    - out_emb: embeddings para palavras de contexto
    - Utiliza produto escalar para scoring de similaridade
    """
    def __init__(self, vocab_size, emb_dim):
        super().__init__()
        self.in_emb = torch.nn.Embedding(vocab_size, emb_dim)
        self.out_emb = torch.nn.Embedding(vocab_size, emb_dim)

    def forward(self, center, context, neg):
        """
        Forward pass do modelo.
        
        Parâmetros:
        ----------
        center : torch.Tensor
            Índices das palavras centrais
        context : torch.Tensor
            Índices das palavras de contexto positivas
        neg : torch.Tensor
            Índices das palavras de contexto negativas
            
        Retorna:
        -------
        tuple
            (scores_positivos, scores_negativos)
        """
        v = self.in_emb(center)
        u_pos = self.out_emb(context)
        u_neg = self.out_emb(neg)

        pos_score = (v * u_pos).sum(dim=1)
        neg_score = torch.bmm(u_neg, v.unsqueeze(2)).squeeze(2)
        return pos_score, neg_score
```

```python
# Treinamento do Word2Vec
def train_word2vec(texts_train, emb_dim=50, window=2, min_count=8, epochs=1,
                   batch_size=4096, neg_k=3, lr=0.03, max_pairs=40000):
    """
    Função principal de treinamento do modelo Word2Vec Skip-gram.
    
    Implementa o pipeline completo de treinamento incluindo:
    1. Construção do vocabulário
    2. Preparação dos dados de treinamento
    3. Inicialização do modelo neural
    4. Loop de treinamento com otimização Adam
    5. Negative Sampling com distribuição de frequência suavizada
    
    Parâmetros:
    ----------
    texts_train : list
        Textos de treinamento
    emb_dim : int
        Dimensionalidade dos vetores de embedding
    window : int
        Tamanho da janela de contexto
    min_count : int
        Frequência mínima para inclusão no vocabulário
    epochs : int
        Número de épocas de treinamento
    batch_size : int
        Tamanho do batch para processamento
    neg_k : int
        Número de exemplos negativos por exemplo positivo
    lr : float
        Taxa de aprendizado para otimização Adam
    max_pairs : int
        Número máximo de pares skip-gram a gerar
        
    Retorna:
    -------
    tuple
        (modelo_treinado, vocabulario, contador)
        
    Justificativa dos Hiperparâmetros:
    - emb_dim=50: Dimensionalidade adequada para vocabulário de domínio específico
    - window=2: Janela pequena para capturar relações sintáticas locais
    - neg_k=3: Balanceamento entre eficiência e qualidade das representações
    - Distribuição p = freq^0.75: Suavização para Negative Sampling (Mikolov et al.)
    """
    stoi, itos, counter = build_vocab(texts_train, min_count=min_count)
    seqs = texts_to_indices(texts_train, stoi)
    pairs = generate_skipgram_pairs(seqs, window=window, max_pairs=max_pairs)

    vocab_size = len(itos)
    model = SkipGramNS(vocab_size, emb_dim)
    opt = torch.optim.Adam(model.parameters(), lr=lr)

    # Distribuição negativa para negative sampling (freq^0.75 suavização)
    freqs = np.array([counter.get(w, 1) for w in itos], dtype=np.float64)
    freqs = freqs ** 0.75
    freqs = freqs / freqs.sum()

    n_pairs = len(pairs)
    for ep in range(epochs):
        idx = np.random.permutation(n_pairs)
        total_loss = 0.0

        for start in range(0, n_pairs, batch_size):
            end = min(start + batch_size, n_pairs)
            batch_idx = idx[start:end]
            batch = pairs[batch_idx]
            
            centers = torch.tensor(batch[:, 0], dtype=torch.long)
            contexts = torch.tensor(batch[:, 1], dtype=torch.long)
            
            neg_samples = np.random.choice(vocab_size, size=(len(batch), neg_k), p=freqs)
            negs = torch.tensor(neg_samples, dtype=torch.long)

            pos_score, neg_score = model(centers, contexts, negs)
            
            pos_loss = torch.nn.functional.logsigmoid(pos_score).sum()
            neg_loss = torch.nn.functional.logsigmoid(-neg_score).sum()
            loss = -(pos_loss + neg_loss)

            opt.zero_grad()
            loss.backward()
            opt.step()
            total_loss += loss.item()

        print(f"Epoch {ep+1}/{epochs} - Loss médio: {total_loss/n_pairs:.4f}")

    return model, stoi, counter
```

```python
texts_train = X_train.tolist()
w2v_model, stoi, counter = train_word2vec(texts_train)
```

```python
# Classificador baseado em médias dos embeddings + LogisticRegression
class SentenceVectorizer(BaseEstimator, TransformerMixin):
    """
    Vetorizador de sentenças baseado em Word2Vec treinado.
    
    Converte sentenças em representações vetoriais através de:
    1. Tokenização simples do texto (split por espaços)
    2. Mapeamento de tokens para embeddings Word2Vec
    3. Agregação por média aritmética dos vetores válidos
    
    Características da Implementação:
    - Herda de BaseEstimator e TransformerMixin para compatibilidade com sklearn
    - Ignora tokens não presentes no vocabulário (OOV handling)
    - Utiliza média aritmética como estratégia de composição
    - Mantém dimensionalidade consistente com modelo Word2Vec base
    """
    def __init__(self, w2v_model, stoi, emb_dim=50):
        self.w2v_model = w2v_model
        self.stoi = stoi
        self.emb_dim = emb_dim

    def fit(self, X, y=None):
        return self

    def transform(self, X):
        """
        Converte lista de sentenças em matriz de vetores.
        
        Processo de Vetorização:
        1. Tokenização simples de cada sentença (split por espaços)
        2. Filtragem de tokens presentes no vocabulário
        3. Extração de embeddings do modelo Word2Vec
        4. Média aritmética para representação da sentença
        5. Vetor zero para sentenças sem tokens válidos
        
        Parâmetros:
        ----------
        X : list
            Lista de strings representando sentenças
            
        Retorna:
        -------
        np.ndarray
            Matriz (n_sentences, emb_dim) com representações vetoriais
        """
        unk = self.stoi.get('<unk>', 1)
        vecs = []
        for doc in X:
            idxs = [self.stoi.get(w, unk) for w in doc.split()]
            if len(idxs) == 0:
                vecs.append(np.zeros(self.emb_dim))
                continue
            emb = self.w2v_model.in_emb(torch.tensor(idxs, dtype=torch.long)).detach().numpy()
            vecs.append(emb.mean(axis=0))
        return np.vstack(vecs)
```

```python
# Pipeline completo: SentenceVectorizer + LogisticRegression
w2v_clf = Pipeline([
    ('sent_vec', SentenceVectorizer(w2v_model, stoi, emb_dim=50)),
    ('clf', LogisticRegression(max_iter=2000))
])
```

```python
# Treinando e avaliando o classificador Word2Vec + LogisticRegression
w2v_clf.fit(X_train, y_train)
pred_w2v = w2v_clf.predict(X_test)
```

```python
# Cálculo do F1 Score
f1_w2v = f1_score(y_test, pred_w2v, average='weighted')
print("\nF1 Score (weighted) - Word2Vec (PyTorch) + LogisticRegression (TESTE):", f1_w2v)
print("\nClassification report:")
print(classification_report(y_test, pred_w2v))
```

## Modelo campeão

```python
# MODELO CAMPEÃO
# Do carregamento da base até a validação final

import pandas as pd
import re

from unidecode import unidecode
from sklearn.model_selection import train_test_split
from sklearn.pipeline import Pipeline
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.svm import LinearSVC
from sklearn.metrics import f1_score, classification_report

# 1) Carregamento da base
df = pd.read_csv('https://dados-ml-pln.s3.sa-east-1.amazonaws.com/tickets_reclamacoes_classificados.csv', delimiter=';')

# 2) Limpeza do texto
def basic_clean(text):
    """
    Realiza o pré-processamento básico de texto para análise de NLP.
    
    Esta função implementa técnicas fundamentais de normalização textual que são
    essenciais para melhorar a qualidade da representação dos dados textuais.
    O objetivo é reduzir a variabilidade superficial do texto mantendo seu
    conteúdo semântico, permitindo que os algoritmos de machine learning
    identifiquem padrões mais consistentes.
    
    Parâmetros:
    ----------
    text : str ou qualquer tipo
        O texto a ser processado. Será convertido para string se necessário.
    
    Retorna:
    -------
    str
        Texto processado e normalizado.
    
    Técnicas aplicadas:
    ------------------
    - Conversão para minúsculas: reduz dimensionalidade do vocabulário
    - Remoção de acentos: normaliza caracteres especiais do português
    - Remoção de URLs: elimina ruído não semântico
    - Remoção de pontuação: foca no conteúdo lexical
    - Normalização de espaços: padroniza separação de tokens
    """
    if not isinstance(text, str):
        text = str(text)
    text = text.lower()
    text = unidecode(text)
    text = re.sub(r"http\S+|www\.\S+", " ", text)
    text = re.sub(r"[^a-z0-9\s]", " ", text)
    text = re.sub(r"\s+", " ", text).strip()
    return text

X = df['descricao_reclamacao'].astype(str).apply(basic_clean)
y = df['categoria'].astype(str)

# 3) Split treino e teste 75%/25%
X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    test_size=0.25,
    random_state=42,
    stratify=y
)
import nltk
nltk.download('stopwords')

# lista de stopwords do NLTK
stops = nltk.corpus.stopwords.words('portuguese')

# 4) Pipeline do modelo
campeao = Pipeline([
    ('tfidf', TfidfVectorizer(
        ngram_range=(1,1),
        min_df=1,
        max_df=0.9,
        stop_words=stops,
        use_idf=False
    )),
    ('clf', LinearSVC(C=0.5))
])

# 5) Treinamento do modelo
campeao.fit(X_train, y_train)

# 6) Avaliação no conjunto de teste
pred = campeao.predict(X_test)

# Cálculo do F1 Score
f1 = f1_score(y_test, pred, average='weighted')
print("F1 Score:", f1)

print("\nClassificação:")
print(classification_report(y_test, pred))
```

```python
# Salvando o modelo
import joblib

joblib.dump(campeao, "modelo_campeao.joblib")
print("Modelo salvo com sucesso!")

# Função de classificação de novos chamados
def classificar_chamado(texto):
    """
    Recebe um texto de chamado e retorna a categoria prevista.
    
    Esta função implementa o pipeline completo de classificação para novos chamados,
    aplicando o mesmo pré-processamento utilizado durante o treinamento e utilizando
    o modelo campeão treinado para fazer a predição da categoria.
    
    Parâmetros:
    ----------
    texto : str
        Texto do chamado a ser classificado
        
    Retorna:
    -------
    str
        Categoria prevista pelo modelo
    
    Processo de Classificação:
    -------------------------
    1. Aplicação da função basic_clean para normalização
    2. Vetorização TF-IDF usando o vocabulário treinado
    3. Classificação usando LinearSVC otimizado
    4. Retorno da categoria com maior probabilidade
    """
    texto_limpo = basic_clean(texto)
    return campeao.predict([texto_limpo])[0]
```

```python
# Teste do modelo com novos textos (simulação real)
exemplos = [
    "Meu cartão foi cobrado duas vezes e preciso de estorno",
    "Gostaria de renegociar meu empréstimo pessoal",
    "Minha conta bancária foi bloqueada sem aviso",
    "Fui vítima de fraude e preciso registrar disputa",
    "Tenho dúvidas sobre financiamento imobiliário"
]

print("Demonstração do modelo campeão em exemplos práticos:")
print("=" * 60)

for texto in exemplos:
    categoria_prevista = classificar_chamado(texto)
    print("Texto:", texto)
    print("Categoria prevista:", categoria_prevista)
    print("-" * 60)
```
