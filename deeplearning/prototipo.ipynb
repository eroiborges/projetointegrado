```python
"""
Protótipo de Rede Neural Feedforward (FNN) para Predição de Indicadores Educacionais
Curso: Deep Learning - FIAP
Objetivo: Prever risco de evasão ou probabilidade de sucesso baseado em dados educacionais

Arquitetura:
- Camada de entrada: 10 features (notas, frequência, tempo de resposta, etc.)
- 2 Camadas ocultas com ativação Sigmoid
- Camada de saída com ativação Sigmoid (classificação binária)
- Otimizador: SGD (taxa de aprendizado 0.01)
- Função de perda: Binary Cross-Entropy
"""

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.optimizers import SGD
from tensorflow.keras.utils import plot_model

# ==============================================================================
# DEFINIÇÃO DOS PARÂMETROS DA REDE
# ==============================================================================

# Número de features de entrada
# Exemplos: notas em diferentes disciplinas, frequência, tempo de resposta,
# número de acessos à plataforma, tempo gasto em atividades, etc.
input_features = 10

print("=== CONSTRUINDO PROTÓTIPO FNN PARA INDICADORES EDUCACIONAIS ===")
print(f"Número de features de entrada: {input_features}")

# ==============================================================================
# DEFINIÇÃO DA ARQUITETURA FNN
# ==============================================================================

model = Sequential([
    # Camada Oculta 1: 16 neurônios com ativação Sigmoid
    Dense(16, activation='sigmoid', input_shape=(input_features,), name='hidden_layer_1'),
    
    # Camada Oculta 2: 8 neurônios com ativação Sigmoid  
    Dense(8, activation='sigmoid', name='hidden_layer_2'),
    
    # Camada de Saída: 1 neurônio com ativação Sigmoid (classificação binária)
    # Saída: probabilidade de sucesso/risco de evasão (0-1)
    Dense(1, activation='sigmoid', name='output_layer')
])

print("\n✓ Arquitetura FNN definida com sucesso!")

# ==============================================================================
# COMPILAÇÃO DO MODELO
# ==============================================================================

# Configuração conforme especificações do projeto
model.compile(
    optimizer=SGD(learning_rate=0.01),  # Stochastic Gradient Descent
    loss='binary_crossentropy',         # Função de perda para classificação binária
    metrics=['accuracy']                # Métrica de avaliação
)

print(" Modelo compilado com SGD (lr=0.01) e Binary Cross-Entropy")

# ==============================================================================
# EXIBIÇÃO DO RESUMO DA ARQUITETURA
# ==============================================================================


print("RESUMO DA ARQUITETURA FNN")
model.summary()

# ==============================================================================
# GERAÇÃO DO DIAGRAMA DA ARQUITETURA
# ==============================================================================

try:
    # Gerar e salvar diagrama da arquitetura conforme especificado
    plot_model(
        model, 
        to_file='fnn_architecture.png',  # Nome conforme requisitos
        show_shapes=True,                # Mostrar dimensões das camadas
        show_layer_names=True,           # Mostrar nomes das camadas
        rankdir='TB',                    # Orientação top-bottom
        dpi=150                          # Qualidade da imagem
    )
    print("\n Diagrama da arquitetura salvo como 'fnn_architecture.png'")
    
except Exception as e:
    print(f"\n Erro ao gerar diagrama: {e}")
    print("Verifique se graphviz está instalado no sistema")


print("PROTÓTIPO FNN CRIADO COM SUCESSO!")

print("Características do modelo:")
print(f"• Features de entrada: {input_features}")
print("• Camadas ocultas: 2 (16 e 8 neurônios)")
print("• Ativação: Sigmoid (todas as camadas)")
print("• Otimizador: SGD (lr=0.01)")
print("• Função de perda: Binary Cross-Entropy")
print("• Aplicação: Predição de indicadores educacionais")
```

# Protótipo FNN para Indicadores Educacionais

## Resumo do Projeto

Este protótipo implementa uma **Rede Neural Feedforward (FNN)** para predição de indicadores educacionais, como risco de evasão ou probabilidade de sucesso acadêmico.

## Arquitetura Implementada

- **Entrada**: 10 features (notas, frequência, tempo de resposta, etc.)
- **Camadas Ocultas**: 2 camadas com 16 e 8 neurônios respectivamente
- **Ativação**: Sigmoid em todas as camadas
- **Saída**: 1 neurônio com ativação Sigmoid (probabilidade 0-1)
- **Otimizador**: SGD com taxa de aprendizado 0.01
- **Função de Perda**: Binary Cross-Entropy

- **Total de Parâmetros**: 321 (1.25 KB)
